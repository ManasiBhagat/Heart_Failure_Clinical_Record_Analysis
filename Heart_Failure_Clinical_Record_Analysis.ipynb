```python
# Coverting Dataset into Array
new_dataset=np.array(data_set[0:],dtype=np.object)
new_dataset
```

```python
# ARRAY INDEXING(1D AND 2D)
print(data_set)
arr_index_2D=new_dataset[25,8]
print("Gender of the Patient:",arr_index_2D)
arr_index_1D=np.array(new_dataset[25],dtype=None)
print("Detail of the Patient:",arr_index_1D)
```

```python
#Array Slicing for fetching 1st-20 patient record
arrslice = new_dataset[0:20,0:]
print(arrslice)
```

```python
#Array reshapping to see individual patient detail
arrreshape=np.reshape(new_dataset,new_dataset.shape+(1,))
print("Details of the individual patient are as follows")
print(arrreshape)
```

```python
#Array Iteration:
Average=new_dataset[0:,1:2]
avgLength = len(Average)
sum = 0
for avg in range(0,avgLength):
  sum += Average[avg]
actualAverage =sum/avgLength
print("Average Age=",int(actualAverage))
```

```python
# Array join
arrjoin=np.array([['P40',55,1,231,0,0,276000,1.8,1,0,1],['P41',65,0,230,1,1,150000,1.0,0,1,0],['P42',65,0,230,1,1,150000,1.0,0,1,0]])
newdata= np.vstack((data_set,arrjoin))
print("New Patient Record",(newdata))
```

```python
#Array Splitting:
split_Array=np.vsplit(newdata,6) 
print(split_Array)
```

```python
platelets = np.array(data_set['platelets'])
print(platelets)
```

```python
#Array filter
filter_Array =[]
for i in platelets:
  if i < 300000:
    filter_Array.append(True)
  else:
    filter_Array.append(False)
platelet_new = platelets[filter_Array]
print(platelet_new)
pltStatus=np.array(filter_Array)
print(pltStatus)


```
